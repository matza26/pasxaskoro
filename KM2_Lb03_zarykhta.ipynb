## Задание 3.1. Анимированное построение линии на плоскости


### Задание 3.1a

а) $\color{red}{\text{Создайте анимацию}}$ построения графика некоторой
явно заданной аналитической функции $y = y(x)$ по значениям $x$,
последовательно изменяющимся от $x_{\min}$ до $x_{\max}$ с шагом
step. Выполните задание для $y(x) = \sin(x)^{4}$, $x_{\min} = 1$, $x_{\max} = 5$, $step = 0.05$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib widget
```

```python
def y(x):
 return np.sin(x)**4
```

```python
x_min = 1.; x_max = 5.; step = 0.05
```

```python
x_array = np.arange(x_min, x_max, step)
y_array = y(x_array)
```

```python
matrix = np.transpose([x_array, y_array])
```

```python
fig1 = plt.figure()
```

```python
y_min, y_max = np.min(y_array), np.max(y_array)
```

```python
ax1 = plt.axes()
plt.axis([x_min-0.5, x_max+0.5, y_min-0.5, y_max+0.5])
```

```python
line1, = ax1.plot([],[],'b')
print(type(line1))
line1.get_xdata(), line1.get_ydata()
```

```python
def at_frame1(point):
    x_coord = list(line1.get_xdata())
    y_coord = list(line1.get_ydata())

    x_coord.append(point[0])
    y_coord.append(point[1])

    line1.set_data(x_coord, y_coord)
```

```python
help(at_frame1)
```

```python
fig1 = plt.figure()
ax1 = plt.axes()
plt.axis([x_min-0.5, x_max+0.5, y_min-0.5, y_max+0.5])
line1, = ax1.plot([], [],'b')
FuncAnimation(fig1, at_frame1, frames=matrix, repeat=False, interval=80)
```

```python
plt.close('all')
```

### Задание 3.1b

```python
def x1(t):
    return np.log(t + np.sqrt(t ** 2 + 1))
```

```python
def y1(t):
    return t * np.sqrt(t ** 2 + 1)
```

```python
t_min = 1.
t_max = 5.
step = (t_max - t_min) / 100
```

```python
t_array = np.arange(t_min, t_max, step)
```

```python
x1_array = x1(t_array)
```

```python
y1_array = y1(t_array)
```

```python
matrix1 = np.transpose([x1_array, y1_array])
```

```python
x1_max, x1_min, y1_min, y1_max = np.max(x1_array), np.min(x1_array), np.min(y1_array), np.max(y1_array)
```

```python
plt.axis([x1_min-0.5, x1_max+0.5, y1_min-10, y1_max+0.5])
```
